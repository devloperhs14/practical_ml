# Data Exploration

* Data Exploration refers to diving into loaded data to get a feel of what it contains.
* Its the initial step in data analysis
* It uncovers characterstics, patterns and potential problems in loaded data (often messy)

# Loading Modules & Data

```python
from sklearn.datasets import load_diabetes
import pandas as pd
```

```python
diabetes = load_diabetes(as_frame=True)
```

```python
df = diabetes.data
```

Attribute Information:

- age: age in years
- sex: gender
- bmi: body mass index
- bp: average blood pressure
- s1: tc, total serum cholesterol
- s2: ldl, low-density lipoproteins
- s3: hdl, high-density lipoproteins
- s4: tch, total cholesterol / HDL
- s5: ltg, possibly log of serum triglycerides level
- s6: glu, blood sugar level

## Get Summary Statistics

```python
df.describe()
```

OUTLIER DETECTION

* Upper - If the 25th percentile and min are far apart, outliers might be present

* Lower - If the 75th percentile and max are far apart, outliers might be present


```python
# get specific summary stats - user defined percentile
df.describe(percentiles=[0.05, 0.25, 0.75, 0.95])
```

```python
?df.describe
```

```python
# include only specific dtype
df.describe(include=['float64'])
```

```python
# ignore specific dtypes
df.describe(exclude=['float64']) # error as no object type is present
```

```python
# transpose the summary stats (col-> rows, summary stats -> cols)
df.describe().T
```
